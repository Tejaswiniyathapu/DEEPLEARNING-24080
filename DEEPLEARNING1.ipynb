```python
import numpy as np

# -----------------------------
# AND Gate Dataset
# -----------------------------
X = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])

y = np.array([0, 0, 0, 1])

# -----------------------------
# Perceptron Training Function
# -----------------------------
def train_perceptron(X, y, lr=0.1, epochs=10):

    # Initialize weights and bias
    weights = np.zeros(X.shape[1])
    bias = 0

    # Training Loop
    for epoch in range(epochs):

        for i in range(len(X)):

            # Step 1: Linear Combination
            linear_output = np.dot(X[i], weights) + bias

            # Step 2: Activation Function
            y_pred = 1 if linear_output >= 0 else 0

            # Step 3: Calculate Error Update
            update = lr * (y[i] - y_pred)

            # Step 4: Update Weights and Bias
            weights += update * X[i]
            bias += update

    return weights, bias

# -----------------------------
# Train Model
# -----------------------------
weights, bias = train_perceptron(X, y)

print("Final Weights :", weights)
print("Final Bias    :", bias)

# -----------------------------
# Prediction Function
# -----------------------------
def predict(X, weights, bias):

    predictions = []

    for sample in X:

        linear_output = np.dot(sample, weights) + bias

        prediction = 1 if linear_output >= 0 else 0

        predictions.append(prediction)

    return np.array(predictions)

# -----------------------------
# Test Predictions
# -----------------------------
predictions = predict(X, weights, bias)

print("\nActual Outputs    :", y)
print("Predicted Outputs :", predictions)

# -----------------------------
# Accuracy
# -----------------------------
accuracy = np.mean(predictions == y) * 100

print("\nAccuracy =", accuracy, "%")

# -----------------------------
# Decision Boundary
# -----------------------------
print("\nDecision Boundary Equation:")

print(f"{weights[0]:.2f}x1 + {weights[1]:.2f}x2 + ({bias:.2f}) = 0")
```

```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam

X = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
], dtype=np.float32)

y = np.array([
    [0],
    [1],
    [1],
    [0]
], dtype=np.float32)

model = Sequential()
model.add(Dense(8, input_shape=(2,), activation='relu'))

model.add(Dense(4, activation='relu'))

model.add(Dense(1, activation='sigmoid'))

model.compile(
    optimizer=Adam(learning_rate=0.01),
    loss='binary_crossentropy',
    metrics=['accuracy']
)

history = model.fit(
    X,
    y,
    epochs=1000,
    verbose=0
)

loss, accuracy = model.evaluate(X, y, verbose=0)

print("Final Loss:", round(loss,4))
print("Final Accuracy:", round(accuracy*100,2), "%")

predictions = model.predict(X)

print("\nPredictions")
print("-------------------------------")

for i in range(len(X)):
    predicted = 1 if predictions[i][0] >= 0.5 else 0
    print(f"Input: {X[i].astype(int)}")
    print(f"Predicted: {predicted}")
    print(f"Actual: {int(y[i][0])}")
    print("-------------------------------")
```

```python
import numpy as np
import pandas as pd
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam

# XOR Dataset
X = np.array([
    [0,0],
    [0,1],
    [1,0],
    [1,1]
], dtype=np.float32)

y = np.array([[0],[1],[1],[0]], dtype=np.float32)

experiments = [
    ("sigmoid", 0.01),
    ("sigmoid", 0.1),
    ("relu", 0.01),
    ("relu", 0.1)
]

results = []

for activation, lr in experiments:

    model = Sequential([
        Dense(4, input_shape=(2,), activation=activation),
        Dense(1, activation="sigmoid")
    ])

    model.compile(
        optimizer=Adam(learning_rate=lr),
        loss="binary_crossentropy",
        metrics=["accuracy"]
    )

    history = model.fit(
        X,
        y,
        epochs=100,
        verbose=0
    )

    loss, acc = model.evaluate(X, y, verbose=0)

    results.append([
        activation.capitalize(),
        lr,
        round(loss,4),
        round(acc*100,2)
    ])

df = pd.DataFrame(
    results,
    columns=[
        "Activation",
        "Learning Rate",
        "Final Loss",
        "Final Accuracy (%)"
    ]
)

print(df)
```

```python
import numpy as np
import pandas as pd

from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Dropout
from tensorflow.keras.optimizers import Adam

# Load MNIST
(X_train, y_train), (X_test, y_test) = mnist.load_data()

X_train = X_train.astype("float32") / 255
X_test = X_test.astype("float32") / 255

experiments = [
    (0.001,128,32,0.2),
    (0.001,256,64,0.3),
    (0.0005,256,128,0.5),
    (0.01,128,64,0.2),
    (0.001,512,32,0.4)
]

results=[]

for lr,hidden,batch,drop in experiments:

    model = Sequential([
        Flatten(input_shape=(28,28)),
        Dense(hidden,activation='relu'),
        Dropout(drop),
        Dense(10,activation='softmax')
    ])

    model.compile(
        optimizer=Adam(learning_rate=lr),
        loss='sparse_categorical_crossentropy',
        metrics=['accuracy']
    )

    history=model.fit(
        X_train,
        y_train,
        epochs=5,
        batch_size=batch,
        validation_split=0.2,
        verbose=0
    )

    train_loss=history.history['loss'][-1]
    val_acc=history.history['val_accuracy'][-1]

    overfit="No"

    if history.history['accuracy'][-1]-val_acc>0.03:
        overfit="Yes"

    results.append([
        f"LR={lr}, Hidden={hidden}, Batch={batch}, Dropout={drop}",
        round(train_loss,4),
        round(val_acc*100,2),
        overfit
    ])

df=pd.DataFrame(results,
columns=[
"Hyperparameters",
"Training Loss",
"Validation Accuracy",
"Overfitting?"
])

print(df)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.metrics import confusion_matrix, classification_report

# -------------------------------
# Evaluate the Model
# -------------------------------
loss, accuracy = model.evaluate(X_test, y_test, verbose=0)

print("Test Accuracy:", round(accuracy * 100, 2), "%")

# -------------------------------
# Predict Test Data
# -------------------------------
predictions = model.predict(X_test)

predicted_labels = np.argmax(predictions, axis=1)

# -------------------------------
# Confusion Matrix
# -------------------------------
cm = confusion_matrix(y_test, predicted_labels)

plt.figure(figsize=(8,6))

sns.heatmap(
    cm,
    annot=True,
    fmt='d',
    cmap='Blues',
    xticklabels=range(10),
    yticklabels=range(10)
)

plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.title("Confusion Matrix - MNIST")

plt.show()

# -------------------------------
# Classification Report
# -------------------------------
print("\nClassification Report\n")

print(classification_report(y_test, predicted_labels))
```
